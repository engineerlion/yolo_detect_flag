```python
import os
from tqdm import tqdm
root = '/data/web_crawl'
img_dir = os.listdir(root)

for idx,d in enumerate(tqdm(img_dir)):
    img_dirr = os.path.join(root,d)
    img_names = os.listdir(img_dirr)
    for im in img_names:
        img_path = os.path.join(img_dirr,im)
        new_name = str(idx)+'_'+im
        #print(new_name)
        img_newpath = os.path.join(img_dirr,new_name)
        #print(img_path,'\n',img_newpath)
        
        os.rename(img_path,img_newpath)
        #assert 0
        
```

```python
mvout_dir = '/data/web_crawl/map_dataset/'
root = '/data/web_crawl'
img_dir = os.listdir(root)

for idx,d in enumerate(tqdm(img_dir)):
    img_dirr = os.path.join(root,d,'*')
    commd = 'mv {} {}'.format(img_dirr,mvout_dir)
    print(commd)
    os.system(commd)
    #assert 0
```

```python
import cv2
data_dir = '/data/web_crawl/map_dataset/'
save_dir = '/data/web_crawl/map_dataset_clean/'
imgs = os.listdir(data_dir)
error_list = []
for im in tqdm(imgs):
    im_path = os.path.join(data_dir,im)
    save_path = os.path.join(save_dir,im)
    #print(im_path)
    try:
        image = cv2.imread(im_path)
        image = cv2.resize(image,(640,480))
    #print(image.shape)
        cv2.imwrite(save_path,image)
    except:
        error_list.append(im)
    #assert 0
```

```python
from imagededup.methods import PHash
phasher = PHash()
root = '/data/web_crawl/map_dataset_clean'
# Generate encodings for all images in an image directory

encodings = phasher.encode_images(image_dir=root)

```

```python
# Find duplicates using the generated encodings
duplicates = phasher.find_duplicates(encoding_map=encodings)
```

```python
from imagededup.utils import plot_duplicates
plot_duplicates(image_dir=root,
                duplicate_map=duplicates,
                filename='25_000447.jpg')
```
